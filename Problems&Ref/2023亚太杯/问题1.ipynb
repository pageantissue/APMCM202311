中国新能源电动汽车发展受到多方面因素的影响，这些因素涵盖了政策、技术、市场和社会等多个方面。为了分析这些因素，我们可以建立一个数学模型，考虑以下关键因素：

1. **政策因素：** 政府的政策对新能源汽车市场有着直接的影响。可以考虑制定的补贴政策、减排目标以及相关法规的力度等。用一个政策强度的参数来表示政府支持的力度，例如 \( P \)。

2. **技术因素：** 新能源汽车的技术水平和成本是影响市场竞争力的关键。可以使用技术创新指数来表示新技术的引入和应用速度，例如 \( T \)。

3. **经济因素：** 汽车经济发展的好坏是影响新能源汽车发展的重要因素。居民消费指数可以用来表示经济因素，例如 \( E \)。

3. **资源环境因素：** 市场需求、竞争状况和消费者接受度是影响新能源汽车发展的重要因素。市场份额、销售增长率等指标可以用来表示市场因素，例如 \( R \)。

4. **社会因素：** 包括消费者对环保的态度、社会文化对新能源汽车的认同程度等。可以使用社会因素指数来表示社会对新能源汽车的接受程度，例如 \( S \)。

基于以上因素，我们可以建立一个简化的数学模型：

$ Development = P \times T \times E \times R \times S$

其中，\( Development \) 表示新能源汽车的发展水平。这个模型假设各个因素是相互独立的，实际情况可能更为复杂，可以考虑引入相互影响的交叉项。

数学模型的具体形式可以根据实际情况和数据的可用性进行调整。这个模型可以用来分析各个因素对新能源汽车发展的相对影响，并为政府、企业等决策者提供参考，以制定更有针对性的政策和战略。

### 以下是本文构建的指标体系

=

1. 政策相关指标。根据以往经验表明，早期新兴产业的发展受政策青睐的影响， 在收益较慢的绿色产业更能体现。新能源汽车虽然符合当年社会和消费者的需要， 但在初期的表现并不乐观，更何况要在成熟完备的传统汽车市场占据一席之地，因 此政府出台有利政策对消费者而言将有最直接的诱惑力，其中包括各种补贴及提供 配套设施等，这些措施既可以增加新能源汽车市场的竞争优势，配套设施又可以为 暂时弥补新能源汽车落后于传统汽车的特性，为新能源汽车发展保驾护航。然而政 策相关指标量化存在困难，同时含有较强的主观性，因此本文选取新能源汽车政策 数、公共充电桩保有量两个指标衡量政策因素对新能源汽车销量的影响。 
2. 技术相关指标。随着新能源汽车的发展，新能源汽车产业由政策指引转变为 技术驱动，由技术而产生的消费刺激对新能源汽车有更深重的影响，可以预见，未 来各企业为了占据新能源汽车市场，将通过寻求技术进步从而提高新能源汽车销量。 但是考虑到新能源汽车品牌众多，技术水平参差不齐，相关指标难以获得且体系繁 多，数据相对隐蔽，而专利是新能源汽车行业技术进步主要表现。因此可以基于专 业信息了解新能源汽车技术创新状况，所以本文选择中国新能源汽车专利申请量指 标。在汽车行业发展中，虽然有各种优惠政策，但相同性能下传统汽车更加便宜， 相同价格下，传统汽车性能更好，因此本文也将选择传统汽车销量作为传统汽车技 术水平的评估指标。从消费者角度考虑，新能源汽车的焦点为里程焦虑和环保价值， 而里程问题主要取决于动力电池发展水平，且对于新能源汽车的电池回收是新能源 汽车环保的主要亮点，因此相关动力电池企业注册量、电池回收企业注册量也作为 新能源汽车技术指标。 
3. 经济相关指标。经济水平往往体现出居民的消费意愿，以宏观经济水平、投 资环境水平等方面，选取具体指标评价我国新能源汽车产业发展前景。国内生产总 值是衡量国家和地区经济总量最常用的指标，由于国内生产总值受到人口总数的影 响，因此本文选取实际人均国内生产总值指数来衡量国民经济的宏观水平；居民消 费水平代表了国民消费能力，可以反映居民的消费意愿；居民收入往往决定了生活 质量，而由于城乡发展水平不均衡，因此选择城镇居民消费水平和农村居民消费水 平加以区分。除此之外，近年来相关投资事件和投资金额迅速增加，作为新兴产业， 该行业的发展很大一部分是由资金支持所决定的，因此本文选择新能源汽车投融资 事件数、新能源汽车产业投融资披露总额作为投资环境指标。 
4. 资源环境相关指标。资源环境包括自然资源和环境资源，前者主要指可以为 汽车产业提供的原材料或者是能源，后者指汽车保有量适度规模的条件。对于传统 汽车而言，汽油和柴油为主要动力来源，具有稀缺性，因此原油产量和汽油价格影 响传统汽车发展，从而影响了新能源汽车发展，因此本文选取原油产量和汽油价格 作为自然环境指标。除此之外，交通布局结构也将限制汽车产业的发展，通过对交 通状况的改善从而可以刺激交通需求，从而刺激汽车产业的发展。换言之，基础设 施的建设对我国新能源汽车发展也将存在影响，因此本文选择道路面积、道路长度、 作为我国基础交通设施建设指标。而公共交通工具是社会中必不可少的交通工具， 其便捷程度可以直接决定大众的出行方式，因此每万人拥有公共交通车辆也是研究 新能源汽车销量影响因素指标之一。新能源汽车相较于传统汽车二氧化碳排放量大 大减少，因此碳排放也成为选择购买新能源汽车的可能因素。 
5. 社会效益相关指标。作为新能源汽车的购买者和使用者，研究新能源汽车销 量的影响因素必须考虑到人口因素，主要包括人口总量、人口素质和人口结构因素。 对于有汽车购买需求的人来说，拥有汽车驾驶证是最基本的条件，因此以汽车驾驶 员人数来衡量消费群体。对于新兴产业，国家虽然花费大量人力物力资源，但依旧 没有被消费者普遍接受，主要原因是消费者对其认知不够，对其价格、性能等方面 缺乏详细的了解。通过调查可以看到，早期购买新能源汽车的消费者，大多是具有 环保意识的高学历的年轻人，因此本文选取老龄化率、高等教育毛入学率作为人口 素质指标。新能源汽车的试行主要是从城市开始的，一方面城市的交通需求比较大， 增加了新能源汽车的机会，另一方面，为缓解交通压力和污染问题，传统汽车出行 有很多限制，这也为新能源汽车提供了最大的竞争优势，除此之外，城市的充电基 础设施比较完备，运营环境良好，因此选择城市化率作为指标之一。 因此，为综合全面的分析我国新能源汽车销量的影响因素，以相关文献及理论 基础作为支撑，结合考虑数据的可获得性，本文将从政策、经济、技术、资源环境、 社会效益五个方面出发，共选取 22 个因素作为我国新能源汽车的年度销量候选影响 因素，将我国新能源汽车年度销量作为因变量，相关变量描述如表所示。

### 选用AHP评价模型作为指标体系权重的构建，由于大部分指标数据是缺失的，在构建专佳评价矩阵时，将主要参考其他文献的指标体系权重结果构建专家评价矩阵，在本文中，主要参考的是基于灰色关联在客观数据上的建模得到的指标结果

```python
import numpy as np
class AHP:
  """
  相关信息的传入和准备
  """
 
  def __init__(self, array):
    ## 记录矩阵相关信息
    self.array = array
    ## 记录矩阵大小
    self.n = array.shape[0]
    # 初始化RI值，用于一致性检验
    self.RI_list = [0, 0, 0.52, 0.89, 1.12, 1.26, 1.36, 1.41, 1.46, 1.49, 1.52, 1.54, 1.56, 1.58,1.59]
    # 矩阵的特征值和特征向量
    self.eig_val, self.eig_vector = np.linalg.eig(self.array)
    # 矩阵的最大特征值
    self.max_eig_val = np.max(self.eig_val)
    # 矩阵最大特征值对应的特征向量
    self.max_eig_vector = self.eig_vector[:, np.argmax(self.eig_val)].real
    # 矩阵的一致性指标CI
    self.CI_val = (self.max_eig_val - self.n) / (self.n - 1)
    # 矩阵的一致性比例CR
    self.CR_val = self.CI_val / (self.RI_list[self.n - 1])
 
  """
  一致性判断
  """
 
  def test_consist(self):
    # 打印矩阵的一致性指标CI和一致性比例CR
    print("判断矩阵的CI值为：" + str(self.CI_val))
    print("判断矩阵的CR值为：" + str(self.CR_val))
    # 进行一致性检验判断
    if self.n == 2: # 当只有两个子因素的情况
      print("仅包含两个子因素，不存在一致性问题")
    else:
      if self.CR_val < 0.1: # CR值小于0.1，可以通过一致性检验
        print("判断矩阵的CR值为" + str(self.CR_val) + ",通过一致性检验")
        return True
      else: # CR值大于0.1, 一致性检验不通过
        print("判断矩阵的CR值为" + str(self.CR_val) + "未通过一致性检验")
        return False
 
  """
  算术平均法求权重
  """
 
  def cal_weight_by_arithmetic_method(self):
    # 求矩阵的每列的和
    col_sum = np.sum(self.array, axis=0)
    # 将判断矩阵按照列归一化
    array_normed = self.array / col_sum
    # 计算权重向量
    array_weight = np.sum(array_normed, axis=1) / self.n
    # 打印权重向量
    print("算术平均法计算得到的权重向量为：\n", array_weight)
    # 返回权重向量的值
    return array_weight
 
  """
  几何平均法求权重
  """
 
  def cal_weight__by_geometric_method(self):
    # 求矩阵的每列的积
    col_product = np.product(self.array, axis=0)
    # 将得到的积向量的每个分量进行开n次方
    array_power = np.power(col_product, 1 / self.n)
    # 将列向量归一化
    array_weight = array_power / np.sum(array_power)
    # 打印权重向量
    print("几何平均法计算得到的权重向量为：\n", array_weight)
    # 返回权重向量的值
    return array_weight
 
  """
  特征值法求权重
  """
 
  def cal_weight__by_eigenvalue_method(self):
    # 将矩阵最大特征值对应的特征向量进行归一化处理就得到了权重
    array_weight = self.max_eig_vector / np.sum(self.max_eig_vector)
    # 打印权重向量
    print("特征值法计算得到的权重向量为：\n", array_weight)
    # 返回权重向量的值
    return array_weight
def matrix_reverse(m):
    for i in range(len(m)):
        for o in range(len(m)):
            if m[i][o] == 0:
                m[i][o] = 1/m[o][i]
                
    return m
```
